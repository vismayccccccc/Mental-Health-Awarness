```python
#Step 3 Visulaization
```

```python
#1 Age Distribution
```

```python
plt.hist(df['Age'], bins=10)
plt.title("Age Distribution of Respondents")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#2 Gender Distribution
```

```python
df['Gender'].value_counts().head(10).plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
#3 Family History VS Mental Health
```

```python
sns.countplot(x='family_history', data=df)
plt.title("Family History of Mental Health Issues")
plt.show()
```

```python
#4 Treatment Taken
```

```python
sns.countplot(x='treatment', data=df)
plt.title("People Who Sought Mental Health Treatment")
plt.show()
```

```python
#5 Work Interference
```

```python
sns.countplot(x='work_interfere', data=df)
plt.title("Mental Health Interfering with Work")
plt.show()
```

```python
# 6 Company Benefits Awareness
```

```python
sns.countplot(x='benefits', data=df)
plt.title("Awareness of Mental Health Benefits at Workplace")
plt.show()
```

```python
#15 Correlation Heatmap
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
